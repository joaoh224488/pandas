```python
import pandas as pd
df = pd.read_csv('pokemon_data.csv')
```

# Ordenação e descrição dos dados
### describe()

Para descrever os dados do ponto de vista estatístico, a forma mais fácil e imediata é por meio da função describe:

```python
df.describe()
```

* count é a quantidade de elementos de cada coluna;
* mean é a média e é o valor "esperado" da distribuição. Em média, você esperaria obter esse número;
* std é o 'standard deviation' é o desvio padrão do conjunto de dados. O tamanho das oscilações dos dados em relação à média. Especificamente, é o desvio quadrado médio da média;
* min é o valor mínimo;
* 50% é a mediana e sua diferença da média fornece informações sobre a inclinação da distribuição. É também outra definição de média que é robusta para valores discrepantes nos dados;
* 25% e 75% são o primeiro e o terceiro quartil, respectivamente, dão perspectiva sobre a curtose(o achatamento da curva da função de distribuição de probabilidade).

### Ordenação de dados
A função sort funciona de forma análoga ao sort de listas padrão em python:

```python
df.sort_values('Name') # Ordem alfábetica para strings
```

```python
df.sort_values('HP') # Ordem número para números
```

Enquanto nas listas o sort pode sofrer um reverse= True para inverter a ordem, no pandas, visando ao mesmo objetivo, usaremos ascending = False.

```python
df.sort_values('Name', ascending= False)
```

Também podemos pegar apenas colunas específicas do dataframe e decidir a ordem:

```python
df.sort_values(['HP', 'Type 1'], ascending=[True, False])
# Observe que o dataframe gerado está ordenado pelo HP primariamente e secundariamente e só depois pela ordme alfabética do tipo
```
