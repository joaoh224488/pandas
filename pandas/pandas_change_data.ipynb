```python
import pandas as pd
import instaloader
df = pd.read_csv('pokemon_data.csv')
```

# Realizando mudanças no dataframe:
### Adicionando coluna

```python
# Coluna 'Total' que soma todos os stats do pokémon da row
df['Total'] = df['Attack'] + df['HP'] + df['Defense'] + df['Speed'] + df['Sp. Def'] + df['Sp. Atk']
df.head(5)
```

Para adicionar uma coluna usando inloc:

```python
df['Total'] = df.iloc[:, 4:10].sum(axis= 1) # axis = 0 --> soma horizontal; axis = 1 --> soma vertical
#O slicing do iloc é lido como: ['todas as células', colunas 4 à 9]
df.head()
```

### Removendo uma coluna com drop:

```python
df = df.drop(columns=['Total'])
df
```

### Movendo as colunas:


```python
cols = list(df.columns)
df = df[cols[:4] + [cols[-1]] + cols[4:12]]
df.head(5)
```

### Salvando dados

```python
df
```

```python
df.to_csv('new_pkmn_data.csv', index= False)
# Adiciona o CSV no diretório
# index= False serve para que o index padrão do Pandas (esse 0,1,2,3... que apaerece do lado é não é a pokedéx)
# não seja salvo no novo arquivo csv
```

```python
# Podemos salvar em excel tambèm:

df.to_excel('new_pkmn_data.xlsx', index= False)
```
