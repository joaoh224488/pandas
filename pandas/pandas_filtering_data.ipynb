```python
import pandas as pd

df = pd.read_csv('new_pkmn_data.csv')
df
```

### Filtrando os dados:

```python
# mostrar apenas pokÃ©mons tipo grass e poison

df[(df['Type 1'] ==  'Grass') & (df['Type 2'] == 'Poison')] # por algum motivo, se usar 'and' ao invÃ©s do '&' dÃ¡ erro
```

#### Agora para fazer uma lista de pokÃ©mons que  possuem o Type 1 Dragon **ou** Fire

```python
df[(df['Type 1'] == 'Dragon') | (df['Type 1'] == 'Fire')] # ao invÃ©s do or, usamos o |
# o | fica junto com a contrabarra no teclado
```

```python
# Dataframe de pokÃ©mons Dragon ou Fire com Total > 300

df[(df['Type 1'] == 'Dragon') & (df['Total'] >= 600) | (df['Type 1'] == 'Fire') & (df['Total'] >= 600)]
```

```python
pseudolendarios = df[(df['Total'] >= 600) & (df['Legendary'] == False)]

pseudolendarios.reset_index(drop = True, inplace = True) # NÃ£o esqueÃ§a de resetar o index sempre que gerar um novo df
pseudolendarios
```

```python
# Ainda fica com a coluna do index salva no df novo
# vamos tirÃ¡-la e tirar tambÃ©m a coluna 'Legendary' que tÃ¡ sobrando agora
pseudolendarios = pseudolendarios.drop(columns=['Legendary']) 
pseudolendarios
```

```python
# Para retirar os megapokÃ©mons da lista:

pseudolendarios.loc[~pseudolendarios['Name'].str.contains('Mega')]

# O sinal de til '~' aÃ­ identifica negaÃ§Ã£o. Ã‰ equivalente ao 'not'

# LÃª-se: pseudolendarios que nÃ£o contÃªm 'mega' no nome.
```
