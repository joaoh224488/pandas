```python
import pandas as pd

df = pd.read_csv('new_pkmn_data.csv')
df
```

### Filtrando os dados:

```python
# mostrar apenas pokémons tipo grass e poison

df[(df['Type 1'] ==  'Grass') & (df['Type 2'] == 'Poison')] # por algum motivo, se usar 'and' ao invés do '&' dá erro
```

#### Agora para fazer uma lista de pokémons que  possuem o Type 1 Dragon **ou** Fire

```python
df[(df['Type 1'] == 'Dragon') | (df['Type 1'] == 'Fire')] # ao invés do or, usamos o |
# o | fica junto com a contrabarra no teclado
```

```python
# Dataframe de pokémons Dragon ou Fire com Total > 300

df[(df['Type 1'] == 'Dragon') & (df['Total'] >= 600) | (df['Type 1'] == 'Fire') & (df['Total'] >= 600)]
```

```python
pseudolendarios = df[(df['Total'] >= 600) & (df['Legendary'] == False)]

pseudolendarios.reset_index(drop = True, inplace = True) # Não esqueça de resetar o index sempre que gerar um novo df
pseudolendarios
```

```python
# Ainda fica com a coluna do index salva no df novo
# vamos tirá-la e tirar também a coluna 'Legendary' que tá sobrando agora
pseudolendarios = pseudolendarios.drop(columns=['Legendary']) 
pseudolendarios
```

```python
# Para retirar os megapokémons da lista:

pseudolendarios.loc[~pseudolendarios['Name'].str.contains('Mega')]

# O sinal de til '~' aí identifica negação. É equivalente ao 'not'

# Lê-se: pseudolendarios que não contêm 'mega' no nome.
```
