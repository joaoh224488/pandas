```python
import pandas as pd
from pycep_correios import get_address_from_cep, WebService, exceptions
```

```python
address = get_address_from_cep('03931-010', webservice=WebService.CORREIOS)
address
```

```python
df = pd.read_excel('Pasta de trabalho.xlsx')
df
```

```python
df['cep']
```

```python
ceps = []
for i in range(len(df['cep'])):
    cep_string = str(df['cep'][i])
    if len(cep_string) >= 7:
        if len(cep_string) == 7:
            ceps.append(f'0{cep_string[0:4]}-{cep_string[4:]}')
        elif len(cep_string) == 8:
            ceps.append(f'{cep_string[0:4]}-{cep_string[4:]}')
        elif len(cep_string) == 9:
            ceps.append(cep_string)
    else:
        ceps.append(None)
ceps
```

```python
logradouro = []

for i in range(len(ceps)):
    if type(ceps[i]) == str:
        try:
            address = get_address_from_cep(ceps[i], webservice=WebService.CORREIOS)
            logradouro.append(address)
            
        except exceptions.InvalidCEP as eic:
            logradouro.append(None)

        except exceptions.CEPNotFound as ecnf:
            logradouro.append(None)

        except exceptions.ConnectionError as errc:
           logradouro.append(None)

        except exceptions.Timeout as errt:
            logradouro.append(None)

        except exceptions.HTTPError as errh:
            logradouro.append(None)

        except exceptions.BaseException as e:
            logradouro.append(None)
    else:
        logradouro.append(ceps[i])
```

```python
df['cep']= ceps
```

```python
ruas = []
for i in logradouro:
    if i != None:
        if i['logradouro'] == None:
            ruas.append(i['cidade'])
        else:
            ruas.append(i['logradouro'])
    else:
        ruas.append(None)
ruas
```

```python
df['logradouro'] = ruas
df
```

```python
df.to_excel('Pasta de trabalho.xlsx', index= False)
```
